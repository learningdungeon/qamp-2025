### ðŸŽ¯ Objective
Simulate a system composed of two independent bits and observe the possible measurement outcomes.

### ðŸ§  Context
When multiple systems are independent, the result of measuring one does not affect the other.  
This task reinforces how combined systems behave when there is **no correlation** between them.

### ðŸ“Œ Instructions
1. Define two independent bits with their own probabilities.
2. Measure both bits repeatedly.
3. Combine the results into two-bit outcomes.
4. Count and print how often each outcome appears.


```python
# =========================
# CODING TASK 3
# Two Independent Bits
# =========================

import random  # Library used to generate random values

# Define the state of each bit
# Bit A is in a probabilistic state (superposition-like)
bit_A = {
    "0": 0.5,
    "1": 0.5
}

# Bit B is always measured as "0"
bit_B = {
    "0": 1.0,
    "1": 0.0
}

# Number of measurements
shots = 1000

# Dictionary to count two-bit outcomes
results = {
    "00": 0,
    "01": 0,
    "10": 0,
    "11": 0
}

# Perform repeated measurements
for _ in range(shots):
    # Measure bit A
    a = "0" if random.random() < bit_A["0"] else "1"
    
    # Measure bit B
    b = "0" if random.random() < bit_B["0"] else "1"
    
    # Combine the results and count them
    results[a + b] += 1

# Print the measurement results
print("Measurement results:", results)


"""
EXPECTED RESULT (explanation):

When you run this program, you should see results similar to:

Measurement results: {'00': 504, '01': 0, '10': 496, '11': 0}

Key observations:
- Outcomes '01' and '11' never appear because bit B is always '0'.
- Outcomes '00' and '10' appear with similar frequencies.
- Each bit is measured independently of the other.

"""
```
