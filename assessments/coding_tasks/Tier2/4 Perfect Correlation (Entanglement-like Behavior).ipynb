###  Objective
Simulate a two-bit system where the measurement outcomes are perfectly correlated.

###  Context
In some systems, the results of measurements are not independent.  
Instead, the outcomes are **linked**: measuring one bit determines the result of the other.

This behavior prepares us to understand **quantum entanglement**, without using quantum hardware yet.

###  Instructions
1. Simulate a system that can only produce the outcomes `00` or `11`.
2. Each outcome should occur with the same probability.
3. Repeat the experiment many times.
4. Count and print the measurement results.


```python
# =========================
# CODING TASK 4
# Perfect Correlation (Entanglement-like Behavior)
# =========================

import random  # Library used to generate random values

# Number of measurements
shots = 1000

# Dictionary to count the allowed outcomes
results = {
    "00": 0,
    "11": 0
}

# Perform repeated measurements
for _ in range(shots):
    # Randomly choose one of the two correlated outcomes
    if random.random() < 0.5:
        results["00"] += 1
    else:
        results["11"] += 1

# Print the measurement results
print("Measurement results:", results)


"""
EXPECTED RESULT (explanation):

When you run this program, you should see results similar to:

Measurement results: {'00': 503, '11': 497}

Important observations:
- The outcomes '01' and '10' never occur.
- Measuring the first bit immediately tells you the value of the second.
- The system cannot be described as two independent bits.

"""
```
