###  Objective
Simulate a system that is in a probabilistic state, where each possible outcome has a defined probability.

###  Context
Instead of choosing outcomes with equal probability, we now define a **state** using probabilities.  
This idea prepares us to understand how quantum states behave *before measurement*.

###  Instructions
1. Define a state using probabilities for outcomes `0` and `1`.
2. Simulate measurements based on those probabilities.
3. Repeat the experiment many times.
4. Count and print the measurement results.


```python
# =========================
# CODING TASK 2
# Probabilistic State Simulation
# =========================

import random  # Library used to generate random values

# Define a probabilistic state
# 50% chance of measuring "0"
# 50% chance of measuring "1"
state = {
    "0": 0.5,
    "1": 0.5
}

# Number of measurements
shots = 1000

# Dictionary to count the results
results = {
    "0": 0,
    "1": 0
}

# Perform repeated measurements
for _ in range(shots):
    # Generate a random number between 0 and 1
    r = random.random()
    
    # Compare it with the probability of measuring "0"
    if r < state["0"]:
        results["0"] += 1
    else:
        results["1"] += 1

# Print the measurement results
print("Measurement results:", results)


"""
EXPECTED RESULT (explanation):

When you run this program, you should see results similar to:

Measurement results: {'0': 497, '1': 503}

The exact values may vary each time you execute the code.
However:
- The outcome frequencies reflect the probabilities defined
  in the state.
- Changing the probabilities in the 'state' dictionary
  will directly affect the measurement results.

"""
```
