###  Objective
Simulate repeated measurements of a bit that can take the values `0` or `1` at random.

###  Context
In computing (and later in quantum computing), a **measurement** is a process that produces an observable result.  
In this task, we use Python to simulate many measurements and count how often each outcome appears.

###  Instructions
1. Generate random measurements of a bit (`0` or `1`).
2. Repeat the experiment many times.
3. Count how many times each result appears.
4. Print the final results.


```python
# =========================
# CODING TASK 1
# Basic Bit Measurement Simulation
# =========================

import random  # Library used to generate random values

# Number of times the measurement is repeated
shots = 1000

# Dictionary to count the measurement results
results = {
    "0": 0,
    "1": 0
}

# Repeat the experiment 'shots' times
for _ in range(shots):
    # Randomly choose between "0" and "1"
    bit = random.choice(["0", "1"])
    
    # Increase the corresponding counter
    results[bit] += 1

# Print the final measurement results
print("Measurement results:", results)


"""
EXPECTED RESULT (explanation):

When you run this program, you should see something like:

Measurement results: {'0': 508, '1': 492}

The exact numbers may change each time you run the code,
but in general:
- The number of '0' and '1' outcomes will be similar.
- The larger the value of 'shots', the closer the results
  will be to a 50/50 distribution.
"""
```
