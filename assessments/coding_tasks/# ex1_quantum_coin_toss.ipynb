```python
# ex1_quantum_coin_toss.py
"""
Exercise 1: Quantum Coin Toss
Learn: Superposition, measurement, basic statistics
Updated: February 2026
"""

import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

print("ğŸª EXERCISE 1: QUANTUM COIN TOSS")
print("=" * 50)

# Create circuit: 1 qubit, 1 classical bit
qc = QuantumCircuit(1, 1)

# Apply Hadamard gate (creates superposition)
qc.h(0)

# Measure
qc.measure(0, 0)

print("\nğŸ“ Your quantum circuit:")
print(qc.draw(output='text'))

# Run simulation
print("\nâš¡ Flipping quantum coin 100 times...")

simulator = AerSimulator()
from qiskit import transpile
compiled_qc = transpile(qc, simulator)

# Run 100 shots (coin tosses)
job = simulator.run(compiled_qc, shots=100)
result = job.result()
counts = result.get_counts()

print(f"\nğŸ“Š Results:")
print(f"  Heads (0): {counts.get('0', 0)} times")
print(f"  Tails (1): {counts.get('1', 0)} times")

# Calculate percentages
total = sum(counts.values())
heads_pct = (counts.get('0', 0) / total) * 100
tails_pct = (counts.get('1', 0) / total) * 100

print(f"\nğŸ“ˆ Percentages:")
print(f"  Heads: {heads_pct:.1f}%")
print(f"  Tails: {tails_pct:.1f}%")

# Expected: Roughly 50/50
expected_ratio = 50
deviation = abs(heads_pct - expected_ratio)

print(f"\nğŸ¯ Deviation from 50/50: {deviation:.1f}%")
if deviation < 10:
    print("âœ… Great! Quantum coin is fair (approximately)")
else:
    print("âš ï¸  Hmm, that's quite skewed. Try running more shots!")

# Visualize
print("\nğŸ“ˆ Generating histogram...")
fig = plot_histogram(counts, title="Quantum Coin Toss Results")
plt.show()

# Challenge questions
print("\n" + "=" * 50)
print("ğŸ¤” THINK DEEPER:")
print("1. Why isn't it exactly 50/50?")
print("2. What happens if you run 1000 shots instead of 100?")
print("3. Try removing the H gate. What happens?")
print("4. What's the difference between quantum randomness and")
print("   classical (computer) randomness?")
```
