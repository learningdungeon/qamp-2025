### Question 1

What is the primary difference in the **Hamiltonian** between QAOA and VQE for Quantum Chemistry?

A) In QAOA, the Hamiltonian represents a graph problem we invented; in VQE, it represents a physical system given by nature.

B) In QAOA, the Hamiltonian is non-Hermitian; in VQE, it is Hermitian.

C) QAOA uses a Mixer Hamiltonian, but VQE does not use any Hamiltonian.

D) VQE Hamiltonians are always diagonal, while QAOA Hamiltonians are not.

**Answer & Explanation:**

**A)** is correct. QAOA solves combinatorial problems (like Max-Cut) where we design the cost function $H_C$. In VQE for chemistry, the Hamiltonian $H$ is the physical energy operator of the molecule (Schr√∂dinger equation), determined by nature.

### Question 2

Electrons are **fermions**, meaning they obey the Pauli Exclusion Principle. What mathematical property must their wavefunction $\Psi(r_1, r_2)$ satisfy?

A) Symmetry: $\Psi(r_1, r_2) = \Psi(r_2, r_1)$

B) Anti-symmetry: $\Psi(r_1, r_2) = -\Psi(r_2, r_1)$

C) Normality: $\langle \Psi | \Psi \rangle = 0$

D) Linearity: $\Psi(r_1 + r_2) = \Psi(r_1) + \Psi(r_2)$

**Answer & Explanation:**

**B)** is correct. Fermionic wavefunctions must be anti-symmetric under particle exchange. Swapping two electrons introduces a global phase of -1. This is the key physical property that makes simulating them with qubits (which don't naturally have this property) challenging.

### Question 3

In the **Jordan-Wigner Mapping**, what does the state of qubit $j$, $|1\rangle_j$, represent?

A) Orbital $j$ is empty.

B) Orbital $j$ is occupied by an electron.

C) Orbital $j$ has spin UP.

D) Orbital $j$ has spin DOWN.

**Answer & Explanation:**

**B)** is correct. The Jordan-Wigner mapping is an occupancy mapping. $|1\rangle_j$ means "occupied" and $|0\rangle_j$ means "empty". Spin is handled by assigning different indices to spin-up and spin-down orbitals.

### Question 4

The Jordan-Wigner mapping transforms the fermionic annihilation operator $a_j$ into a tensor product of Pauli matrices. Why is there a string of $Z$ operators, $\bigotimes_{k=0}^{j-1} Z_k$, attached to the operator?

A) To measure the energy of the previous orbitals.

B) To calculate the **parity** of the occupied orbitals below $j$, enforcing the anti-symmetry requirement.

C) To initialize the qubits in the ground state.

D) It is an error correction code.

**Answer & Explanation:**

**B)** is correct. Since qubits commute ($X_1 X_2 = X_2 X_1$) but fermions anti-commute ($a_1 a_2 = -a_2 a_1$), we need a non-local operation to mimic the "minus sign" that appears when particles are swapped. The Z-string counts the parity of electrons "below" orbital $j$ to apply this phase correctly.

### Question 5

Consider a system with 4 spin-orbitals mapped to 4 qubits. How would the creation operator $a^\dagger_2$ (create electron in orbital 2) look in the Jordan-Wigner mapping?

A) $\sigma^+_2$

B) $\sigma^+_2 \otimes Z_1 \otimes Z_0$

C) $\sigma^+_2 \otimes X_1 \otimes X_0$

D) $Z_3 \otimes Z_2 \otimes \sigma^+_2$

**Answer & Explanation:**

**B)** is correct. The creation operator $a^\dagger_j$ maps to $(\bigotimes_{k=0}^{j-1} Z_k) \otimes \sigma^+_j$. For $j=2$, the Z-string acts on indices $0$ to $1$. So we get $\sigma^+_2 Z_1 Z_0$.

### Question 6

What is a **Heuristic Ansatz** (also called Hardware Efficient Ansatz)?

A) A circuit derived exactly from the chemistry of the molecule (like UCCSD).

B) A generic circuit structure (layers of rotations and entanglers) designed to be easy to run on quantum hardware, without specific chemical knowledge.

C) A circuit that only uses Clifford gates.

D) A circuit that prepares the Hartree-Fock state.

**Answer & Explanation:**

**B)** is correct. Heuristic ansatz (like `EfficientSU2`) are designed to be shallow and use native gates to survive noise on NISQ devices. They don't encode chemical intuition but rely on the optimizer to find the solution in their broad search space.

### Question 7

What is the main **disadvantage** of using the **UCCSD** (Unitary Coupled Cluster Singles and Doubles) ansatz?

A) It is chemically inaccurate.

B) It cannot preserve the number of electrons.

C) It generates very deep circuits with many CNOT gates, making it sensitive to noise on current hardware.

D) It requires more qubits than the Jordan-Wigner mapping.

**Answer & Explanation:**

**C)** is correct. While UCCSD is chemically accurate and physically motivated, implementing the exponentiated double-excitation operators ($e^{T_2}$) requires decomposing them into long chains of CNOTs and rotations. This depth often exceeds the coherence time of current NISQ devices.

### Question 8

The **Variational Principle** states that for any parameterized state $|\psi(\theta)\rangle$, the expectation value of the Hamiltonian $\langle \psi(\theta) | H | \psi(\theta) \rangle$ satisfies what condition?

A) It is always equal to the ground state energy $E_0$.

B) It is always less than $E_0$.

C) It is always greater than or equal to $E_0$.

D) It is always zero.

**Answer & Explanation:**

**C)** is correct. The expectation value is an upper bound on the true ground state energy. This is the foundation of VQE: by minimizing this expectation value (the cost), we get closer and closer to the true ground state energy $E_0$.

### Question 9

In the VQE hybrid loop, what is the role of the **Estimator** primitive?

A) To update the parameters $\vec{\theta}$.

B) To calculate the energy expectation value $\langle H \rangle$ given a circuit and observable.

C) To return the bitstring with the highest probability.

D) To map the fermions to qubits.

**Answer & Explanation:**

**B)** is correct. The `Estimator` takes the ansatz circuit and the Hamiltonian (as a sum of Pauli strings) and returns the energy value. This value is then fed to the classical optimizer.

### Question 10

When simulating the **Lithium Hydride (LiH)** molecule, why do we generate a curve of Energy vs. Interatomic Distance?

A) To find the ionization energy of the molecule.

B) To find the **bond length** (equilibrium geometry), which corresponds to the minimum of the energy curve.

C) To determine the number of electrons in the system.

D) To test if the Jordan-Wigner mapping is correct.

**Answer & Explanation:**

**B)** is correct. Molecules naturally settle into the geometry that minimizes their total energy. By calculating the ground state energy at different distances, we find the minimum point of the curve, which tells us the stable bond length of the molecule.
