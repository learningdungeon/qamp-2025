### Question 1

What is the primary objective of the BB84 protocol?

A) To send a specific message (like 'Hello World') faster than light.

B) To securely distribute a shared random secret key between two parties.

C) To clone quantum states for backup purposes.

D) To encrypt a message using a public key.

**Answer & Explanation:**

**B)** is correct. The goal of Quantum Key Distribution (QKD) protocols like BB84 is to establish a shared random key for later encryption. Quantum mechanics does not allow faster-than-light communication, and the No-Cloning Theorem explicitly forbids creating perfect copies of unknown quantum states.

### Question 2

In the BB84 protocol, if Alice wants to encode the bit '0' using the **Diagonal (X) Basis**, which quantum state does she send?

A) $|0\rangle$

B) $|1\rangle$

C) $|+\rangle$

D) $|-\rangle$

**Answer & Explanation:**

**C)** is correct. In the X-basis, $|+\rangle$ represents bit '0' and $|-\rangle$ represents bit '1'. The states $|0\rangle$ and $|1\rangle$ correspond to the Rectilinear (Z) Basis.

### Question 3

If Alice sends a qubit in the state $|0\rangle$ (Z-basis) and Bob measures it in the **X-basis**, what will result from Bob's measurement?

A) He will definitely measure '0'.

B) He will definitely measure '1'.

C) He will obtain a random result (0 or 1) with 50% probability each.

D) The qubit will be destroyed without giving a result.

**Answer & Explanation:**

**C)** is correct. The bases are mutually unbiased. Measuring a Z-basis state in the X-basis yields a completely random outcome ($|\langle + | 0 \rangle|^2 = 1/2$).

### Question 4

During the transmission phase, how does Bob decide which basis to use for measuring each incoming qubit?

A) He uses the same basis Alice used.

B) He chooses the basis completely randomly.

C) He always alternates between Z and X.

D) He measures in both bases simultaneously.

**Answer & Explanation:**

**B)** is correct. Bob has no information about the incoming qubit, so he must guess the basis randomly. He cannot know Alice's basis choice until after the transmission is complete.

### Question 5

What happens during the **Sifting** phase of the protocol?

A) Alice and Bob publicly reveal their data bits (0s and 1s).

B) Alice and Bob compare their basis choices and keep bits where the bases matched.

C) Bob tries to guess Alice's basis choices.

D) They use error correction to fix flipped bits.

**Answer & Explanation:**

**B)** is correct. They discard bits where bases mismatched (which produces noise) and keep the ones where they aligned. This forms the 'Sifted Key'.

### Question 6

Which fundamental principle prevents Eve from simply copying the qubit Alice sends, storing the copy, and measuring it later after the bases are revealed?

A) The No-Cloning Theorem.

B) The Superposition Principle.

C) The Pauli Exclusion Principle.

D) The Photoelectric Effect.

**Answer & Explanation:**

**A)** is correct. The No-Cloning Theorem proves it is impossible to create an identical copy of an arbitrary unknown quantum state.

### Question 7

If Eve performs an **Intercept-Resend Attack** on every qubit, what is the expected error rate (QBER) introduced into the sifted key?

A) 0%

B) 11%

C) 25%

D) 50%

**Answer & Explanation:**

**C)** is correct. Eve picks the wrong basis 50% of the time, and Bob gets the wrong result 50% of *those* times. $0.5 \times 0.5 = 0.25$ or 25%.

### Question 8

What is the specific goal of **Error Reconciliation** (e.g., using the Cascade protocol)?

A) To increase the length of the key.

B) To ensure Alice and Bob possess identical keys by correcting mismatches.

C) To determine if Eve is listening.

D) To encrypt the message using the raw key.

**Answer & Explanation:**

**B)** is correct. Noise and eavesdropping cause errors. Reconciliation fixes them so both parties have the same bitstring.

### Question 9

Even after error correction, Eve might know some information about the key. What process is used to reduce Eve's information to effectively zero?

A) Sifting

B) Privacy Amplification

C) Quantum Teleportation

D) Basis Rotation

**Answer & Explanation:**

**B)** is correct. This step compresses the key using hashing (like XORing subsets) to shrink Eve's partial knowledge to near zero.

### Question 10

In the context of the Cascade protocol for error reconciliation, what is the purpose of **parity checks**?

A) To measure the spin of the electrons.

B) To identify blocks of bits that contain an odd number of errors.

C) To encrypt the key before sending it.

D) To randomize the basis choices.

**Answer & Explanation:**

**B)** is correct. If the parities of a block differ between Alice and Bob, they know there is an error in that block. Parity checks allow them to locate errors without revealing the specific bits.
