**Superposition and measurement**

> A quantum system can exist in more than one possible state

## 4Ô∏è‚É£ Challenge Statement

Design a **toy, game, or short story** that represents a *toy qubit*:

- The toy must have **two possible outcomes**
- The outcome must be **undefined before observation**
- The act of observing must **change or define** the result

There is no single correct solution.  
Creativity and explanation are more important than precision.

## Note for the Teacher

This challenge is **not about correctness**.

Focus on:
- Student intuition
- The language students use to describe uncertainty
- Their understanding of observation as an active process

Avoid introducing formal quantum terminology too early.  
Let curiosity lead the discussion.
