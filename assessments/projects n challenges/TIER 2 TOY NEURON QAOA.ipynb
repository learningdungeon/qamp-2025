## 2Ô∏è‚É£ Context and Narrative (Entry Point)

In classical machine learning, a **neuron** takes an input, applies **weights**, and produces an output decision.

Now imagine a different kind of neuron.

This neuron does not decide immediately.  
It explores many possibilities at once and gradually **leans toward the best decision**.

This is the idea behind a **quantum toy neuron**.

In this project, you will explore how a **simple quantum circuit** can behave like a neuron that learns which output is more favorable.

---

## 3Ô∏è‚É£ Core Quantum Idea

**Quantum optimization and probabilistic decision-making**

> A quantum system can explore multiple possibilities simultaneously  
> and favor better solutions through optimization.

---

## 4Ô∏è‚É£ Challenge Statement

Design and analyze a **toy quantum neuron** using the ideas behind **QAOA (Quantum Approximate Optimization Algorithm)**.

Your neuron should:

- Have **one input edge**
- Have **two possible outputs**
- Assign different **weights** to each output
- Use a quantum process to favor the better-weighted outcome

You are not required to code yet,  
but your design should be clear enough that it **could be implemented later in Python**.

## 6Ô∏è‚É£ Toy Neuron Structure (Guided Model)

Use the following **conceptual neuron model**:

- **One input edge**
- **Two output edges**
  - Output A with weight **2**
  - Output B with weight **1**

Think of the weights as indicating **preference**, not certainty.

Ask yourself:
- How can a quantum circuit represent these preferences?
- How might angles or phases encode weights?
- Why might the higher-weight output appear more often, but not always?

This structure will later map naturally to a **QAOA-style circuit in Python**.

##  Note for the Teacher

This challenge bridges **classical AI intuition** and **quantum thinking**.

Encourage students to:
- Compare classical and quantum neurons
- Accept uncertainty as a feature, not a flaw
- Focus on *why* optimization works, not just *that* it works

Avoid deep mathematical formalism at this stage.
The goal is conceptual exploration, not mastery.
