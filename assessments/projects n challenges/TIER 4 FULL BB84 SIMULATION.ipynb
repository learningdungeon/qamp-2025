## 6Ô∏è‚É£ BB84 State Preparation and Encoding

Alice prepares single-qubit states chosen from two **conjugate bases**:

- Computational (Z) basis:
  $$
  |0\rangle,\quad |1\rangle
  $$

- Diagonal (X) basis:
  $$
  |+\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}}, \quad
  |-\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}
  $$

Each transmitted qubit is prepared as one of these four states, encoding both:
- a classical bit
- a choice of basis

The key point:  
states from different bases are **non-orthogonal**.

---

## 7Ô∏è‚É£ Measurement, Collapse, and Basis Mismatch

Bob measures each incoming qubit using a randomly chosen basis (Z or X).

If Bob chooses the same basis as Alice, the measurement outcome is deterministic.

If the bases differ, the measurement outcome is probabilistic:

$$
|\langle 0 | + \rangle|^2 = \frac{1}{2}
$$

After transmission, Alice and Bob publicly compare bases and **discard mismatched measurements**.

The remaining bits form the **raw key**.

## 8Ô∏è‚É£ Eavesdropping as a Quantum Operation

Consider an eavesdropper Eve performing an intercept‚Äìresend attack.

Eve measures the qubit in some basis and resends a new state to Bob.

This introduces a disturbance because Eve cannot know Alice‚Äôs basis in advance.

Formally, Eve‚Äôs intervention replaces Alice‚Äôs pure state $|\psi\rangle$ with a mixed state:

$$
\rho = \sum_i p_i |\phi_i\rangle \langle \phi_i|
$$

This disturbance propagates to Bob‚Äôs measurements, producing an **observable error rate**.

## 9Ô∏è‚É£ Error Rates and Detectability

If Eve measures randomly in Z or X bases, she introduces a quantum bit error rate (QBER) of approximately:

$$
\text{QBER} \approx 25\%
$$

Alice and Bob estimate this error rate by sacrificing a subset of their raw key.

A nonzero QBER is not just noise ‚Äî  
it is **evidence of information leakage**.

Security is therefore inferred, not assumed.

In Dirac notation, security is encoded directly in the inner products of states.
