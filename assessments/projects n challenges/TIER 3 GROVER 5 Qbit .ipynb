## 2Ô∏è‚É£ Context and Narrative (Entry Point)

Searching is one of the most common computational tasks.

In a classical computer, searching through $N$ possibilities usually requires checking them one by one.

Quantum computing offers a different approach.

Instead of testing each option individually, a quantum system can **amplify the probability of the correct answer** through controlled interference.

This project explores **Grover‚Äôs algorithm** using a system of **5 qubits**, focusing on *why* it works rather than just *how* to implement it.

## 3Ô∏è‚É£ Core Quantum Idea

> Grover‚Äôs algorithm increases the probability of measuring a marked state

## 6Ô∏è‚É£ Mathematical Structure of the Problem

A system of 5 qubits spans a Hilbert space of dimension:

$$
N = 2^5 = 32
$$

The initial state is an equal superposition of all basis states:

$$
|\psi_0\rangle = \frac{1}{\sqrt{32}} \sum_{x=0}^{31} |x\rangle
$$

The algorithm applies two main operations repeatedly:

- **Oracle**: flips the phase of the marked state $|w\rangle$
- **Diffusion operator**: reflects amplitudes about their average

Each Grover iteration acts as a **rotation in a two-dimensional subspace** spanned by:
- the marked state
- the uniform superposition of unmarked states
