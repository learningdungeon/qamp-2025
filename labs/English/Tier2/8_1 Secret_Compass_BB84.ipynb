# 3 EXPLAIN — Giving Meaning to What We Observed

In the previous section, we did not talk about quantum theory.  
We only **observed patterns**.

---

## 3.1 From arrows to quantum states

The paper arrows were not just arrows.

They represented **quantum states**.

In quantum computing, the smallest unit of information is a **qubit**.

A qubit is not just:
- 0 or 1  

It is a **direction in a space of possibilities**.

---

## 3.2 Two ways to describe a qubit (two bases)

In our experiment, Antonio used two kinds of arrows:

- Vertical arrows (↑ ↓)
- Diagonal arrows (↗ ↙)

In quantum language, these are called **measurement bases**.

### Vertical Basis (Z basis)
- ↑ represents |0⟩
- ↓ represents |1⟩

### Diagonal Basis (X basis)
- ↗ represents |+⟩
- ↙ represents |−⟩

These bases are **not the same**, and that difference matters.

---

## 3.3 Why bases matter

A key rule of quantum physics:

> **A quantum state does not have a definite answer for all questions at once.**

- If a qubit is prepared in the **Z basis**,  
  it has a clear answer for Z measurements.
- If it is measured in the **X basis**,  
  the result is **random**.

This is not noise.  
This is not a mistake.  
This is how nature works.

---

## 3.4 Measurement is a physical action

When Charlie observed the arrow, something important happened.

He did not:
- reveal a hidden value  

He did:
- **force the system to choose a value**

This is called **measurement**.

> Measurement is not passive.  
> Measurement changes the state.

That is why:
- choosing the “wrong” basis creates randomness
- observing early affects what comes later

---

## 3.5 Why Adara’s observation was detectable

When Adara secretly observed the arrow:

- She also performed a measurement
- She also forced a decision
- The qubit was **changed**, even if no one saw her

Later, when Antonio and Charlie compared results:
- extra disagreement appeared
- something invisible had left evidence

This is the core idea behind **quantum security**:
> You cannot observe a quantum system without leaving traces.

---

## 3.6 No cloning, no copying

Another important rule:

> **A quantum state cannot be perfectly copied.**

If Adara tries to:
- copy the arrow
- observe and resend it

She must measure it first —  
and measurement changes the state.

This is not a technological limit.  
It is a **physical law**.

````python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from random import choice

sim = AerSimulator()

# Random preparation
state = choice(['0', '1', '+', '-'])
basis = choice(['Z', 'X'])

qc = QuantumCircuit(1, 1)

# State preparation
if state == '1':
    qc.x(0)
elif state == '+':
    qc.h(0)
elif state == '-':
    qc.x(0)
    qc.h(0)

# Measurement basis
if basis == 'X':
    qc.h(0)

qc.measure(0, 0)

result = sim.run(qc, shots=1000).result()
counts = result.get_counts()

print(qc)
plot_histogram(counts)
````

> What to observe

- Same basis → predictable result
- Different basis → random outcomes
- Randomness is not noise — it is structural

### Experiment 2 — Hidden observation (intervention)

**Physical idea**

Adara may observe the compass *before* Charlie receives it.  
Her measurement does not need to change anything — just looking can affect outcomes.

**Quantum idea**

- Adara performs a measurement
- Measurement **collapses the quantum state**
- Charlie receives the qubit in the new state
- Even if Charlie uses the same basis as Antonio, results may **disagree**

**Circuit logic**

- Antonio prepares a state
- Adara may measure (random basis)
- Charlie measures (chosen basis)

**Circuit structure (conceptual)**

- Qubit prepared
- Measurement by Adara (optional)
- Collapsed state continues to Charlie
- Measurement by Charlie

>(This simulates interception.)

**What to observe**

- Compare Charlie's measurement with Antonio's original state (only when bases match)
- **Agreement decreases** when Adara observes
- **Disagreement** appears statistically; single events do not reveal intervention
- The more often Adara observes, the lower the overall agreement
- This demonstrates that **quantum observation leaves statistical evidence**, even without an active change

**Key insight**

> You cannot detect a single measurement directly.  
> You detect the **pattern of disturbance** over many runs.


````python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from random import choice, random

sim = AerSimulator()

# ----------------------------
# Parameters
# ----------------------------
shots = 1000
p_adara = 0.01  # Probability that Adara observes you may change the value 0 to 1
results = {'agree': 0, 'disagree': 0}

# ----------------------------
# Run the experiment
# ----------------------------
for _ in range(shots):
    # ------------------------
    # 1. Antonio prepares a qubit
    # ------------------------
    state = choice(['0', '1', '+', '-'])  # Random state
    basis_charlie = choice(['Z', 'X'])    # Charlie's measurement basis
    
    qc = QuantumCircuit(1, 1)
    if state == '1':
        qc.x(0)
    elif state == '+':
        qc.h(0)
    elif state == '-':
        qc.x(0)
        qc.h(0)
    
    # ------------------------
    # 2. Adara may observe (randomly)
    # ------------------------
    if random() < p_adara:
        adara_basis = choice(['Z', 'X'])
        if adara_basis == 'X':
            qc.h(0)
        qc.measure(0, 0)
        qc.reset(0)  # Collapse state
    
    # ------------------------
    # 3. Charlie measures
    # ------------------------
    if basis_charlie == 'X':
        qc.h(0)
    qc.measure(0, 0)
    
    # Execute
    result = sim.run(qc, shots=1).result()
    charlie_outcome = list(result.get_counts().keys())[0]
    
    # ------------------------
    # 4. Determine expected outcome if no disturbance
    # ------------------------
    if basis_charlie == 'Z':
        expected = '0' if state in ['0', '+'] else '1'
    else:  # X basis
        expected = '0' if state in ['+', '0'] else '1'
    
    # ------------------------
    # 5. Classify agree / disagree
    # ------------------------
    if charlie_outcome == expected:
        results['agree'] += 1
    else:
        results['disagree'] += 1

print(qc)
# ----------------------------
# 6. Show results
# ----------------------------
print(f"Probability Adara observed: {p_adara}")
print(f"Results over {shots} shots:", results)
plot_histogram(results)
````

>What to observe

- Even when Antonio and Charlie use the same basis, disagreement increases
- Adara does not need to change anything
- Looking is enough

### Experiment 3 — Redundancy and detection (2–3 qubits)

- Antonio sends **two identical qubits** to Charlie
- Charlie measures both qubits
- If no interference:
    - Outcomes are always identical (`00` or `11`)
    - Agreement is maximal
- If interference/noise occurs:
    - Outcomes can differ (`01` or `10`)
    - Discrepancy signals a disturbance
- Redundancy allows us to **detect errors or interception** even if a single qubit seems fine

**Physical idea**

- Antonio sends the same message multiple times.
- Charlie compares results.

**Quantum idea**

- Redundancy does not correct the qubit
- Redundancy reveals disturbance

**Circuit logic**

- Prepare identical states
- Measure each independently
- Compare statistics

````python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from random import choice, random
import matplotlib.pyplot as plt

sim = AerSimulator()
shots = 500  # reducir para rapidez en clase
p_values = [0, 0.25, 0.5, 0.75, 1.0]  # agregamos p=1

# Set up figure
fig, axes = plt.subplots(1, len(p_values), figsize=(20,4), sharey=True)

for idx, p_adara in enumerate(p_values):
    all_outcomes = {'00':0, '01':0, '10':0, '11':0}
    
    for _ in range(shots):
        qc = QuantumCircuit(2,2)
        
        # Antonio prepares two qubits
        states = [choice(['0','1','+','-']) for _ in range(2)]
        for i, state in enumerate(states):
            if state == '1':
                qc.x(i)
            elif state == '+':
                qc.h(i)
            elif state == '-':
                qc.x(i)
                qc.h(i)
        
        # Adara may intervene
        for i in range(2):
            if random() < p_adara:
                adara_basis = choice(['Z','X'])
                if adara_basis == 'X':
                    qc.h(i)
                qc.measure(i,i)
                qc.reset(i)
        
        # Charlie measures both qubits
        qc.measure([0,1],[0,1])
        result = sim.run(qc, shots=1).result()
        outcome = list(result.get_counts().keys())[0]
        all_outcomes[outcome] += 1
    
    # Plot histogram for this p_adara
    labels = list(all_outcomes.keys())
    values = list(all_outcomes.values())
    colors = ['red','green','green','red']
    
    axes[idx].bar(labels, values, color=colors)
    axes[idx].set_title(f'p_adara = {p_adara}')
    axes[idx].set_xlabel('Qubit outcomes')
    for i, v in enumerate(values):
        axes[idx].text(i, v + 5, str(v), ha='center', fontweight='bold')
    
axes[0].set_ylabel('Counts')
plt.suptitle('2-Qubit Outcomes vs Probability of Adara Intervention')
print(qc)
plt.show()

````

>What to observe

- Without intervention → strong correlations
- With interception → correlations weaken
- Redundancy creates evidence, not certainty

**What these circuits teach us Across all three experiments**

- Measurement is a physical process
- Bases define what can be known
- Observation leaves traces
- Randomness can signal interference
- Trust is built statistically
  
**These are not bugs or errors,**
**they are features of quantum reality**

## Experiment 4 — Random Scale Observation

So far, Adara either:
- always observed, or
- did not observe at all

Now we test something more realistic.

---

### Physical idea

Adara flips a coin:
- Heads → she observes the qubit
- Tails → she lets it pass untouched
- Mapping the probabilites from 0% to 100%

No one knows which runs were observed.

---

### Quantum idea

- Measurement is a physical interaction
- It does not leave a visible mark on a single run
- Its effect appears only **in statistics**

---

### Circuit logic

For each run:
1. Antonio prepares a qubit
2. Adara *may* measure it (random choice)
3. Charlie measures it
4. Results are collected over many shots

## Plot the outcome Agreement Parameter ##


````python
import matplotlib.pyplot as plt
import numpy as np
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from random import choice, random

sim = AerSimulator()
shots = 1000  # adjust for resolution
p_values = np.linspace(0.01, 1, 11)  # de 0.0 a 1.0 en pasos de 0.1
agreement_fraction = []

for p_adara in p_values:
    results = {'agree':0, 'disagree':0}
    for _ in range(shots):
        # Antonio prepares
        state = choice(['0', '1', '+', '-'])
        basis_charlie = choice(['Z', 'X'])
        qc = QuantumCircuit(1,1)
        if state == '1':
            qc.x(0)
        elif state == '+':
            qc.h(0)
        elif state == '-':
            qc.x(0)
            qc.h(0)
        
        # Adara may observe
        if random() < p_adara:
            adara_basis = choice(['Z','X'])
            if adara_basis == 'X':
                qc.h(0)
            qc.measure(0,0)
            qc.reset(0)
        
        # Charlie measures
        if basis_charlie == 'X':
            qc.h(0)
        qc.measure(0,0)
        result = sim.run(qc, shots=1).result()
        charlie_outcome = list(result.get_counts().keys())[0]
        
        # Determine expected outcome
        if basis_charlie == 'Z':
            expected = '0' if state in ['0','+'] else '1'
        else:
            expected = '0' if state in ['+', '0'] else '1'
        
        # Classify
        if charlie_outcome == expected:
            results['agree'] += 1
        else:
            results['disagree'] += 1
    
    # Fraction of agree
    agreement_fraction.append(results['agree']/shots)

# Plot curve
plt.figure(figsize=(7,5))
plt.plot(p_values, agreement_fraction, marker='o', linestyle='-', color='blue')
plt.title('Security vs Probability of Adara observing')
plt.xlabel('Probability Adara observes (p_adara)')
plt.ylabel('Security - Agreement Parameter')
plt.grid(True)
plt.ylim(0,1.05)
plt.show()
````

### What to observe

- **With `p_observe = 0.0`**
  - Adara don´t look so Secure is high when measurement bases match

- **With `p_observe = 0.5`**
  - Adara May looks or not, so Security drops partially

- **With `p_observe = 1.0`**
  - Adara looks so it´s Readed so Security drops significantly


There is:

- no single “trigger moment”
- no visible collapse event
- only **statistical evidence**

---

### Key insight

> **Quantum observation is not detected per event.**  
> **It is detected by pattern incompatibility.**

This is why:

- BB84 requires many bits
- redundancy matters
- trust is statistical, not absolute
