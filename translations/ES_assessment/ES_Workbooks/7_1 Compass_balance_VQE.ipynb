## 2 Explore — VQE as a Quantum Compass (Energy and Equilibrium)

In this session, we are **not trying to “choose the best option”**, as we did in QAOA.  
Instead, we study how a **quantum system naturally settles into a stable state**.

The central question is:

> *If we allow a quantum system to freely adjust itself,  
> which state does it naturally converge to?*

This is the core idea behind the **Variational Quantum Eigensolver (VQE)**.

---

### 2.1 From a Classical Compass to a Quantum One

A classical compass placed in a magnetic field:

- Can rotate freely  
- Experiences a torque  
- Eventually **aligns with the field**

It does not “decide” where to point.  
It simply moves toward the configuration with **minimum energy**.

This principle — *systems seek minimum energy* — is universal in physics.

---

### 2.2 A Qubit as a Quantum Compass

A qubit can be interpreted as a **quantum compass**:

- It is not limited to pointing only up or down  
- It can point in **any direction**
- Its orientation is represented on the **Bloch sphere**

A general single-qubit state is written as:

$$
|\psi\rangle =
\cos\left(\frac{\theta}{2}\right)|0\rangle
+ e^{i\phi}\sin\left(\frac{\theta}{2}\right)|1\rangle
$$

Here:
- $\theta$ controls the tilt of the compass  
- $\phi$ controls the phase (rotation around the axis)

These angles define the **orientation** of the quantum compass.

---

### 2.3 The Magnetic Field → The Hamiltonian

In quantum mechanics, the “field” acting on a system is described by a **Hamiltonian**.

For a single quantum compass in a simple magnetic field, we can write:

$$
H = -B Z
$$

Interpretation:
- $Z$ measures whether the qubit points along or against the field  
- The negative sign means **alignment lowers energy**

The Hamiltonian defines how “costly” or “favorable” each orientation is.

---

### 2.4 Energy Expectation Value

A quantum state does not have a single fixed energy.  
Instead, it has an **expected energy**:

$$
E(\theta, \phi) =
\langle \psi(\theta, \phi) | H | \psi(\theta, \phi) \rangle
$$

This value tells us:

- How stable the current orientation is  
- Whether the system can lower its energy by rotating

Some orientations are unstable.  
One orientation corresponds to **minimum energy**.

---

### 2.5 What VQE Actually Does

The Variational Quantum Eigensolver works as follows:

1. Prepare the quantum compass in a chosen orientation  
2. Measure its energy  
3. Slightly change the orientation  
4. Measure again  
5. Repeat the process  
6. **Converge toward the minimum-energy state**

Importantly:

> VQE does **not** solve equations directly.  
> It allows the quantum system to **find its own equilibrium**.

---

### 2.6 Key Conceptual Distinction

**QAOA**
- Stabilizes *decisions*
- Works with discrete choices
- Focuses on probabilities

**VQE**
- Stabilizes *states*
- Works with continuous parameters
- Focuses on energy

---

###  Key advice

> **VQE is not a decision algorithm.**  
> **It is an equilibrium experiment.**

We “shake” the system  
and observe **where it naturally settles**.

## 3 Explain — Formal Hamiltonian Intuition

> *“To understand where a system settles,  
> we must understand what the system is responding to.”*

In quantum mechanics, that guiding influence is called the **Hamiltonian**.

---

### What Is a Hamiltonian?

The **Hamiltonian** is an operator that represents the **energy structure** of a system.

It answers the question:

> *If the system is in this state, how much energy does it have?*

Mathematically, it is written as:

$$
\hat{H}
$$

But conceptually, think of it as:

- A **magnetic field** acting on compasses  
- A **landscape** with hills and valleys  
- A **rule** that rewards some states and penalizes others  

---

### Energy as Preference

Quantum systems naturally prefer **lower energy configurations**.

If a state $|\psi\rangle$ interacts with a Hamiltonian $\hat{H}$,  
the quantity we care about is the **expected energy**:

$$
E = \langle \psi | \hat{H} | \psi \rangle
$$

This value tells us:

- How well the state aligns with the Hamiltonian
- Whether the system is “comfortable” or “strained”

Lower $E$ means **better alignment**.

---

### Compass Analogy Revisited

| Physical System | Quantum System |
|-----------------|----------------|
| Magnetic field  | Hamiltonian $\hat{H}$ |
| Compass needle  | Quantum state $|\psi\rangle$ |
| Alignment       | Low energy |
| Instability     | High energy |

A compass does not calculate angles.  
It **rotates** until torque disappears.

Likewise, a quantum system does not compute solutions.  
It **settles** into states favored by the Hamiltonian.

---

### Important Clarification

At this stage:

- We are **not** solving the Hamiltonian exactly
- We are **not** finding eigenstates analytically
- We are **not** diagonalizing matrices

Instead, we ask a simpler question:

> *Can we prepare states that naturally lower the energy?*

This is the key intuition behind **VQE**.

---

### Why Measurement Matters

Quantum states are probabilistic.

When we measure repeatedly:

- High-energy components appear less often
- Low-energy components dominate statistics

Equilibrium is revealed **through repetition**, not certainty.

---

### Conceptual Takeaway

> The Hamiltonian defines **what the system prefers**.  
> The quantum state explores **how well it can satisfy that preference**.

VQE emerges when we **let the system try**,  
measure how well it did,  
and adjust.

---

### Learning Objective

Understand the Hamiltonian as an **energy guide**,  
not as a problem to be solved directly.

This intuition prepares us to introduce:
- Variational states
- Parameter tuning
- Energy minimization without exact solutions




## 3B - Explain From Energy Landscapes to the Variational Principle

> *“If the system cannot reach the true ground state directly,  
> can it at least move downhill?”*

This question leads to the **Variational Principle**.

---

### The Core Idea

In quantum mechanics, there exists a powerful guarantee:

> **No trial state can have lower energy than the true ground state.**

Formally:

$$
\langle \psi(\boldsymbol{\theta}) | \hat{H} | \psi(\boldsymbol{\theta}) \rangle
\;\ge\;
E_0
$$

Where:

- $\hat{H}$ is the Hamiltonian  
- $E_0$ is the ground state energy  
- $|\psi(\boldsymbol{\theta})\rangle$ is any **trial state**, controlled by parameters $\boldsymbol{\theta}$  

This is the **variational bound**.

---

### What This Means Conceptually

We may not know the exact ground state.

But we *do* know:

- Every guess has an energy
- Some guesses are better than others
- Energy provides a **directional signal**

The system tells us:

> *“This direction lowers energy. Keep going.”*

---

### Energy as a Landscape

Think again in terms of terrain:

- Each quantum state is a point on a landscape
- The Hamiltonian defines the height (energy)
- The ground state is the lowest valley

The variational method does **not** map the entire terrain.

It walks downhill by **testing nearby positions**.

---

### Role of Parameters

We restrict ourselves to a **family of states**:

$$
|\psi(\boldsymbol{\theta})\rangle
$$

These parameters may represent:

- Rotation angles
- Interaction strengths
- Mixing weights

Changing $\boldsymbol{\theta}$ reshapes the state.

We are not solving equations —  
we are **tuning knobs**.

---

### Measurement Closes the Loop

Quantum hardware cannot give energy directly.

Instead, we:

1. Prepare the state $|\psi(\boldsymbol{\theta})\rangle$
2. Measure observables related to $\hat{H}$
3. Estimate the energy statistically
4. Adjust parameters classically
5. Repeat

This forms a **hybrid quantum–classical loop**. (Looks like QAOA)

---

### Why This Is Powerful

- No exact diagonalization required
- No exponential classical memory
- Noise can be tolerated through statistics
- Physical intuition remains central

This is why VQE works even on early hardware.

---

### Connection to the Compass Analogy

| Compass System | VQE System |
|---------------|------------|
| Shake the compass | Prepare a trial state |
| Let it settle | Measure energy |
| Observe direction | Update parameters |
| Repeat | Iterate |

The compass does not compute north.  
It **relaxes** into it.

---

### Conceptual Boundary

Important distinction:

- QAOA: explores **combinatorial structure**
- VQE: explores **physical energy structure**

But both rely on the same principle:

> *Let the system tell us where it wants to go.*

---

### Learning Objective

Understand the **variational principle** as a guarantee:
energy feedback is meaningful and directional.

This prepares us to:
- Build minimal VQE circuits
- Interpret convergence physically
- Compare VQE and QAOA meaningfully




## 4.1 Elaborate (Part 1) — The Simplest Possible VQE: One Qubit, One Field

> *“Before optimizing molecules or graphs,  
> we must learn how a single quantum compass finds its direction.”*

---

### The Physical System

We consider the **simplest quantum system**:

- One qubit  
- One external magnetic field  
- One preferred direction  

This is not a toy example.  
It is the **hydrogen atom of variational algorithms**.

---

### The Hamiltonian

We choose a Hamiltonian that represents a magnetic field along the $z$-axis:

$$
\hat{H} = - Z
$$

Interpretation:

- Measuring $Z$ tells us whether the qubit points **with** or **against** the field
- The system minimizes energy when it aligns with the field

Ground state:

$$
|0\rangle \quad \text{with energy} \quad E = -1
$$

Excited state:

$$
|1\rangle \quad \text{with energy} \quad E = +1
$$

---

### The Variational Question

Instead of preparing $|0\rangle$ directly, we ask:

> *What happens if we only allow the qubit to rotate freely  
> and let energy guide the alignment?*

This is the essence of VQE.

---

### The Trial State (Ansatz)

We define a **parameterized state**:

$$
|\psi(\theta)\rangle = R_y(\theta)\,|0\rangle
$$

Where:

- $\theta$ is a continuous parameter
- $R_y(\theta)$ rotates the qubit on the Bloch sphere
- Each $\theta$ corresponds to a different compass orientation

This is our **search space**.

---

### Energy as a Function

For this system, the expected energy is:

$$
E(\theta) = \langle \psi(\theta) | \hat{H} | \psi(\theta) \rangle
$$

Which simplifies to:

$$
E(\theta) = -\cos(\theta)
$$

This means:

- $\theta = 0$ → minimum energy
- $\theta = \pi$ → maximum energy

The energy landscape is **smooth and continuous**.

---

### Why This Matters

This single-qubit example shows:

- Optimization without equations
- Learning through measurement
- Continuous parameters instead of binary choices
- Physical intuition driving computation

Before solving complex problems,  
we must understand **how energy guides learning**.

---

### Learning Objective

Identify how a **parameterized quantum state** interacts with a Hamiltonian
to produce a measurable energy landscape.

This prepares us to:
- Implement the VQE loop
- Interpret convergence visually
- Compare VQE behavior with QAOA later




## 4️.2 Elaborate (Part 2) — Measuring Energy: The Core of VQE

> *“The qubit does not tell us its angle.  
> It only tells us how much energy it feels.”*

In VQE, we never access the state directly.  
We interact with it **only through measurements**.

This section shows how **energy becomes feedback**.

---

### From State to Measurement

We prepared the trial state:

$$
|\psi(\theta)\rangle = R_y(\theta)\,|0\rangle
$$

Now we must evaluate:

$$
E(\theta) = \langle \psi(\theta) | \hat{H} | \psi(\theta) \rangle
$$

For our Hamiltonian:

$$
\hat{H} = -Z
$$

This means:

- Measure the qubit in the computational basis
- Assign:
  - $|0\rangle \rightarrow Z = +1$
  - $|1\rangle \rightarrow Z = -1$
- Multiply by $-1$

---

### Measurement as Statistics

Quantum mechanics gives probabilities, not answers.

So we must:

1. Measure the qubit many times (shots)
2. Count how often we get 0 or 1
3. Compute the expectation value

This is the **bridge between quantum and classical**.

---

### The Variational Loop (Conceptual)

1. Choose a value of $\theta$
2. Prepare the state
3. Measure the energy
4. Update $\theta$ to reduce energy
5. Repeat until convergence

We now implement **steps 1–3**.

---

### Learning Objective

Understand how **measurement statistics**
produce a numerical energy value,
which later drives classical optimization.



```python
# ============================================
# 4.2 — One-Qubit Energy Measurement (VQE Core)
# ============================================

import numpy as np
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import matplotlib.pyplot as plt

# --- Simulator ---
simulator = AerSimulator()

# --- Parameter ---
theta = 1.4 # trial angle (can be changed)

# --- Build the trial state (NO measurement) ---
qc_state = QuantumCircuit(1)
qc_state.rx(theta, 0)

state = Statevector.from_instruction(qc_state)

display(plot_bloch_multivector(state, title="Trial State on the Bloch Sphere"))

# --- Build measurement circuit ---
qc = QuantumCircuit(1, 1)
qc.ry(theta, 0)
qc.measure(0, 0)

# --- Run the experiment ---
shots = 1024
job = simulator.run(qc, shots=shots)
result = job.result()
counts = result.get_counts()

print("Measurement counts:", counts)

# --- Plot histogram ---
display(plot_histogram(counts, title="Measurement Results"))

# --- Compute expectation value of Z ---
p0 = counts.get('0', 0) / shots
p1 = counts.get('1', 0) / shots

expectation_Z = p0 - p1
energy = -expectation_Z   # H = -Z

print(f"⟨Z⟩ = {expectation_Z:.3f}")
print(f"Energy E(θ) = {energy:.3f}")

```

## 4 Elaborate (Part 3) — Closing the Loop: Classical Optimization and Convergence

> *“The qubit does not learn.  
> The classical computer does.”*

Until now, we measured the energy of **one trial state**.
But VQE is not about a single measurement.

VQE is about **iteration**.

---

### The Variational Feedback Loop

The Variational Quantum Eigensolver works by closing a loop:

1. A **classical optimizer** proposes parameters ($\theta$)
2. A **quantum circuit** prepares the corresponding state
3. Measurements estimate the **energy**
4. The optimizer updates the parameters to reduce energy
5. The loop repeats until convergence

This loop mirrors a physical process:

> A system explores configurations  
> and naturally settles into a **minimum-energy state**.

---

### Why This Is Not Brute Force

Unlike classical search:

- We do **not** test all angles
- We do **not** scan the Bloch sphere

Instead:
- The optimizer follows the **energy gradient**
- The quantum system provides **physical feedback**

This is why VQE scales better than naive search.

---

### What Convergence Means (Physically)

Convergence does **not** mean:
- The qubit stops moving
- The state becomes fixed

It means:
- Changing $\theta$ no longer reduces the measured energy
- The system has reached a **stable equilibrium**

This is the quantum analogue of:
> A compass needle settling in the magnetic field.

---

### Learning Objective

Understand how **classical optimization**
uses **quantum energy measurements**
to guide the system toward a stable ground state.




```python
# =================================================
# 4.3 — Full VQE Loop: Optimization and Convergence
# =================================================

import numpy as np
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# --- Simulator ---
simulator = AerSimulator()

# --- Hamiltonian: H = -Z ---
shots = 1024
energy_history = []

# --- Energy evaluation function ---
def energy(theta):
    theta = theta[0]   # <-- extract scalar

    qc = QuantumCircuit(1, 1)
    qc.ry(theta, 0)
    qc.measure(0, 0)

    result = simulator.run(qc, shots=shots).result()
    counts = result.get_counts()

    p0 = counts.get('0', 0) / shots
    p1 = counts.get('1', 0) / shots

    expectation_Z = p0 - p1
    E = -expectation_Z

    energy_history.append(E)
    return E

# --- Initial guess ---
theta_initial = [2.5]

# --- Classical optimization ---
result = minimize(
    energy,
    theta_initial,
    method="COBYLA",
    tol=1e-3
)

theta_opt = result.x[0]
energy_opt = result.fun

print("Optimal theta:", theta_opt)
print("Minimum energy:", energy_opt)

# --- Plot convergence ---
plt.figure(figsize=(8, 5))
plt.plot(energy_history, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Energy")
plt.title("VQE Convergence: Energy vs Iteration")
plt.grid(True)
plt.show()

```

### Visualizing the Final Quantum State (Statevector)

After the VQE optimization converges, we can reconstruct the final quantum
state using the optimal parameter $\theta^*$.

This allows us to:
- See **where on the Bloch sphere** the state settled
- Connect energy minimization with **physical alignment**
- Confirm that the system relaxed toward the ground state


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

# --- Rebuild the optimized circuit (no measurement!) ---
qc_final = QuantumCircuit(1)
qc_final.ry(theta_opt, 0)

# --- Obtain final statevector ---
final_state = Statevector.from_instruction(qc_final)

print("Final statevector:")
print(final_state)
theta_extracted = qc_final.data[0].operation.params[0]

print(f"The extracted Theta is: {theta_extracted}")
# --- Visualize on the Bloch sphere ---
plot_bloch_multivector(
    final_state,
    title="Final VQE State on the Bloch Sphere"
)
```

## Key Advice (Critical Insight)

The quantum circuit never changes itself.
The classical optimizer learns how to steer it.

>This hybrid nature is the heart of VQE.

>Quantum: explores states

>Classical: decides direction

>Physics: defines the goal (minimum energy)
