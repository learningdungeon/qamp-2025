### Physical Concept
In classical systems, correlation requires direct interaction or a shared medium.  
In quantum systems, **pre-established correlations** (entanglement) allow the *state of one system* to be **reconstructed elsewhere** using a joint operation and classical communication — without physically transporting the system itself.

### Learning Objective
- Intuitively distinguish **interaction** from **correlation**.  
- Understand the roles of a **pair of correlated systems** and a **third carrier system**, preparing the ground for **quantum teleportation**.

## 3 Explain — From Physical Coupling to the Mathematical Model (Three Qubits)

*In this section, we move deliberately from physical intuition to mathematical structure,  
using Dirac notation as the natural language of composite quantum systems.*

---

### 3.1 From a Compass to a Qubit (Single System)

A single compass needle can point in different directions.  
In the quantum description, a **qubit** is a system whose state is written as a vector in a two-dimensional Hilbert space.

We choose a reference basis:

- $|0\rangle$  → reference direction  
- $|1\rangle$  → opposite direction  

A general single-qubit state is written as:

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle
$$

with the normalization condition:

$$
|\alpha|^2 + |\beta|^2 = 1
$$

**Pedagogical emphasis**

- The qubit is in **one state**, not in two states simultaneously.
- The coefficients $\alpha$ and $\beta$ are **probability amplitudes**, not probabilities.
- At this stage, no measurement is involved.

---

### 3.2 Combining Systems: The Tensor Product

When two physical systems are considered together, their joint description is not obtained by addition,  
but by the **tensor product** of their individual state spaces.

For two qubits $A$ and $B$:

$$
|\psi\rangle_{AB} = |\psi\rangle_A \otimes |\psi\rangle_B
$$

From the single-qubit basis $\{|0\rangle, |1\rangle\}$, we construct the **computational basis** for two qubits:

- $|0\rangle_A \otimes |0\rangle_B = |00\rangle$
- $|0\rangle_A \otimes |1\rangle_B = |01\rangle$
- $|1\rangle_A \otimes |0\rangle_B = |10\rangle$
- $|1\rangle_A \otimes |1\rangle_B = |11\rangle$

**Pedagogical emphasis**

- The tensor product **creates a new space**, it does not mix states.
- A two-qubit system lives in a four-dimensional space.
- Each basis vector already refers to **both systems simultaneously**.

---

### 3.3 A First Example of Entanglement: The Bell State

Some two-qubit states cannot be written as a tensor product of single-qubit states.  
These are called **entangled states**.

The simplest and most important example is the Bell state:

$$
|\Phi^+\rangle = \frac{1}{\sqrt{2}} \left( |00\rangle + |11\rangle \right)
$$

This state describes two qubits whose properties are defined **only jointly**.

**Pedagogical emphasis**

- There is no independent state for each qubit.
- The system must be treated as a single object.
- The detailed theory of Bell states will be developed in a separate notebook.

---

### 3.4 Extending to Three Qubits (Three-System Description)

We now introduce a third qubit, forming a composite system of three parts: $A$, $B$, and $C$.

Roles:

- **Qubit A** carries an *unknown state* $|\psi\rangle$
- **Qubits B and C** share a Bell state $|\Phi^+\rangle$

The total three-qubit state is constructed as:

$$
|\Psi\rangle = |\psi\rangle_A \otimes |\Phi^+\rangle_{BC}
$$

This expression already represents a **single state** in a composite Hilbert space.

---

### 3.5 Explicit Expansion of the Three-Qubit State

Substituting the definitions:

$$
|\Psi\rangle =
(\alpha |0\rangle_A + \beta |1\rangle_A)
\otimes
\frac{1}{\sqrt{2}} \left( |00\rangle_{BC} + |11\rangle_{BC} \right)
$$

Using the distributive property of the tensor product:

$$
|\Psi\rangle =
\frac{1}{\sqrt{2}} \Big(
\alpha |0\rangle_A |00\rangle_{BC}
+ \alpha |0\rangle_A |11\rangle_{BC}
+ \beta |1\rangle_A |00\rangle_{BC}
+ \beta |1\rangle_A |11\rangle_{BC}
\Big)
$$

Writing each term as a three-qubit basis state:

$$
|\Psi\rangle =
\frac{1}{\sqrt{2}} \left(
\alpha |000\rangle
+ \alpha |011\rangle
+ \beta |100\rangle
+ \beta |111\rangle
\right)
$$

**Pedagogical emphasis**

- Each term refers to **all three qubits simultaneously**.
- The state lives in an eight-dimensional space.
- The information in $\alpha$ and $\beta$ is distributed across the full system.

---

### 3.6 Teaching Recommendations (Board or Notebook Work)

To reinforce understanding, construct the state step by step:

1. Write the unknown single-qubit state $|\psi\rangle_A$.
2. Write the Bell state $|\Phi^+\rangle_{BC}$.
3. Form the tensor product explicitly.
4. Expand the expression term by term.
5. Read the physical meaning of each basis state.

**Key reminder**

- No measurement has occurred.
- No information has been transmitted.
- The structure of the state has only been reorganized.

---

### Teacher Advise

The tensor product is not merely a mathematical operation.  
It is the formal language that describes **shared physical reality** in quantum systems.

Mastering this structure is essential before introducing  
quantum circuits, measurements, and teleportation.



## 4 Elaborate — From Mathematical Structure to Quantum Circuit (Teleportation)

> *In this phase, teachers translate the mathematical model into a working quantum circuit.  
The goal is not speed, but clarity: each gate represents a physical or logical action already discussed.*


### 4.1 Roles of the Three Qubits in the Circuit

We explicitly assign meanings to each qubit:

- **Qubit 0 (A)** → the *unknown state* to be teleported  
- **Qubit 1 (B)** → part of the entangled pair (shared channel)  
- **Qubit 2 (C)** → the receiver qubit  

> Keep this mapping visible on the board while building the circuit.

---

### 4.2 Step 1 — Prepare the State to Be Teleported

We begin with an arbitrary single-qubit state on qubit A.

Example:
$$
|\psi\rangle = \alpha|0\rangle + \beta|1\rangle
$$

In practice, this can be prepared using a rotation gate.



```python
from qiskit import QuantumCircuit, transpile, ClassicalRegister, QuantumRegister
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import math
import matplotlib.pyplot as plt

# --- CONFIGURATION ---
theta = 1.1 # The angle for the initial state preparation in radians
# ---------------------

### 4.2 Step 1 — Prepare the State 

qc_initial_A = QuantumCircuit(1)
qc_initial_A.ry(theta, 0)
initial_state_A = Statevector(qc_initial_A)

print(f"--- Initial State (Theta = {theta} radians) ---")
print("Initial State Vector of Qubit A (|psi>):", initial_state_A.data)

# Visualize the initial state on a Bloch sphere (Qubit A)
display(plot_bloch_multivector(initial_state_A, title="Initial State of Qubit A"))

### 4.2 Step 2 — Prepare Circuit
qA = QuantumRegister(1, "A")
qB = QuantumRegister(1, "B")
qC = QuantumRegister(1, "C")
classic = ClassicalRegister(2, "Clssc")

qc = QuantumCircuit(qA, qB, qC, classic)

#  4.2 Step 3 — Prepare the State THETA to Be Transfered
#Prepare the arbitrary state on qubit A (using the same theta)
qc.ry(theta, qA)
qc.barrier()

# 4.2 Step 4 — Prepare the State to Be Transfered
# Create Bell pair between qubit B and qubit C
qc.h(qB)
qc.cx(qB, qC)
qc.barrier()

# ### 4.2 Step 5 — Bell Measurement on Qubits A and B
# Set the qbits for mesuring
qc.cx(qA, qC)
qc.h(qA)
qc.barrier()

# ### 4.2 Step 5 measure qubits A and B into the classical register
# Note: When measuring multi-qubit registers to multi-bit registers in Qiskit,
# the mapping qA, to classic 0 and qB to classic 1
qc.measure(qA, classic[0])
qc.measure(qB, classic[1])
qc.barrier()

# Conditional Corrections on Qubit C as it readed.
# Apply X correction to qubit C if classical bit 1 is 1
with qc.if_test((classic[1], 1)):
    qc.x(qC)

# Apply Z correction to qubit C if classical bit 0 is 1
with qc.if_test((classic[0], 1)):
    qc.z(qC)
    
print("\n--- Full Teleportation Circuit ---")



# ### 4.2 Step 6 SIMULATE AND VIEW THE FINAL STATE ===

# Use a statevector simulator to capture the final quantum state exactly
sim_statevector = AerSimulator(method='statevector')

# We add an instruction to save the state at the end of the circuit
qc.save_statevector()

# Transpile and run the statevector simulation
transpiled_state_circ = transpile(qc, sim_statevector)
result_state = sim_statevector.run(transpiled_state_circ).result()

# Get the final state vector of the entire 3-qubit system
final_statevector = result_state.get_statevector(qc)

print("\n--- Final State ---")
# The Bloch sphere visualizes the individual states of all qubits in the 3-qubit vector
print("Bloch Sphere visualization of all final qubits:")
display(plot_bloch_multivector(final_statevector, title="Final State of Qubits (A, B, C)"))

# To keep the plots visible in some environments
plt.show()
qc.draw(output='mpl', plot_barriers=True) 
```

## 5 Evaluate — Evidence, State Reconstruction, and Teaching Reflection

> *This evaluation focuses on verifying teleportation as a process of **state reconstruction**, not particle transfer, using statevectors and Bloch-sphere visualization.*

---

### 5.1 Comparing Initial and Final States (Core Evidence)

During the simulation, two key quantum states are explicitly constructed and visualized:

1. **Initial state of qubit A**
   \[
   |\psi\rangle = \cos\left(\frac{\theta}{2}\right)|0\rangle + \sin\left(\frac{\theta}{2}\right)|1\rangle
   \]

   - Prepared using a rotation gate `Ry(θ)`
   - Visualized on the Bloch sphere
   - Represents the *unknown state* to be teleported

2. **Final state of qubit C**
   - Extracted from the full 3-qubit statevector
   - Visualized using `plot_bloch_multivector`
   - Should match the original orientation of qubit A

> **Key observation:**  
> Although qubit A is measured and no longer exists as a quantum state,  
> qubit C reproduces the same quantum orientation.
