### 3. Refinamientos de Lógica Computacional (Semana 2)

#### A. La compuerta CNOT: Éxito como copiadora clásica

| Concepto               | Explicación                                                                                   | Acción y salida clave de CNOT |
|------------------------|----------------------------------------------------------------------------------------------|-------------------------------|
| Copia clásica          | Cuando el qubit de control está en un estado clásico definido (∣0⟩ o ∣1⟩), la compuerta CNOT copia perfectamente ese estado al qubit objetivo. | CNOT∣00⟩=∣00⟩, CNOT∣10⟩=∣11⟩ (¡Copia clásica perfecta!) |
| Enfoque en compuerta   | Los estudiantes revisan la compuerta X (NOT) como operación base. La CNOT es simplemente un NOT condicional. | La CNOT confirma su función como pieza central para la computación clásica. |

#### B. La compuerta CNOT: Modo de fallo cuántico

| Concepto               | Explicación                                                                                   | Acción y salida clave de CNOT |
|------------------------|----------------------------------------------------------------------------------------------|-------------------------------|
| Entrada en superposición | Cuando la CNOT se aplica a un estado inicial en superposición, como ∣+⟩∣0⟩, la salida **no** es una copia. | CNOT $(\frac{1}{\sqrt{2}}(\vert 0\rangle + \vert 1\rangle))\vert 0\rangle = \frac{1}{\sqrt{2}}(\vert 00\rangle + \vert 11\rangle)$ (El estado de Bell [Φ+⟩) |
| Creación de entrelazamiento | La salida es un solo **sistema cuántico no separable** donde el estado de q0 no puede describirse independientemente de q1. Esto es el fallo matemático de la clonación. | Este estado de Bell es el recurso fundamental necesario para la teleportación y la codificación superdensa. |

### 4. Plan de lección ejemplar: La paradoja de la CNOT

**Módulo:** Interacción multi-qubit  
Esta lección se centra en la prueba matemática del éxito/fallo de la CNOT usando la salida de probabilidades del Composer.

#### Laboratorio Composer: El éxito y el fallo de la CNOT

| Objetivo               | Los estudiantes demostrarán que la compuerta CNOT es una **copiadora clásica perfecta** pero una **copiadora cuántica imperfecta**, confirmando matemáticamente la razón del Teorema de No-Clonación. |

| Recursos requeridos    | IBM Quantum Composer, Hoja de trabajo Nivel 4 (Cálculos en notación de Dirac) |

**Instrucciones paso a paso**

**Parte 1: CNOT como copiadora clásica (éxito)**
1. **Preparación del estado:** Inicializa q0 en ∣1⟩ usando una compuerta **X**. q1 permanece en ∣0⟩. (Estado inicial: ∣10⟩)
2. **Aplicación de CNOT:** Aplica una compuerta CNOT con q0 (control) y q1 (objetivo).
3. **Predicción:** Los estudiantes usan notación de Dirac para predecir el resultado (∣11⟩).
4. **Verificación:** Agrega operaciones de medición. El resultado debe mostrar 100% de probabilidad de medir ∣11⟩, probando una copia clásica perfecta.

**Parte 2: CNOT como generador de entrelazamiento (fallo)**
1. **Preparación del estado:** Inicializa q0 en el estado ∣+⟩ usando una compuerta **H**. q1 permanece en ∣0⟩. (Estado inicial: ∣+⟩∣0⟩)
2. **Aplicación de CNOT:** Aplica la compuerta CNOT nuevamente (sin cambios en la CNOT).
3. **Predicción:** Los estudiantes usan notación de Dirac para predecir el resultado (∣Φ+⟩).
4. **Verificación:** Agrega operaciones de medición. El resultado debe mostrar **50% de probabilidad de ∣00⟩ y 50% de probabilidad de ∣11⟩**.
5. **Conclusión:** El hecho de que ambos qubits estén ligados y colapsen juntos prueba que la CNOT no creó dos estados ∣+⟩ independientes, sino un solo estado de Bell entrelazado.

### 6. Conclusión y siguientes pasos

Este módulo de **Nivel 4, Semana 2** logra la transición de los estudiantes desde la declaración conceptual del Teorema de No-Clonación hacia la **prueba matemática y computacional** del éxito de la compuerta CNOT en estados clásicos y su inevitable fallo en estados de superposición.

El resultado directo es la comprensión del **entrelazamiento** como recurso—un subproducto necesario del fallo de la CNOT para clonar.
