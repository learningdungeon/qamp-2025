Este módulo culmina el estudio del Teorema de No-Clonación. ### 3. Actividad central: Laboratorio de fallo de clonación universal

#### A. La geometría del fallo (esfera de Bloch)

| Concepto               | Acción                                                                                       | Observación clave        |
|------------------------|----------------------------------------------------------------------------------------------|--------------------------|
| Estado arbitrario      | Usar compuertas RY(θ) y P(λ) para mover el qubit de entrada (q0) a un punto aleatorio en el plano X-Y. | La complejidad del estado desconocido es esencial; no es un autovalor de la base de medición de la CNOT. |
| Corrupción por CNOT    | Aplicar la compuerta CNOT (Control q0, Objetivo q1).                                         | Cuando q0 está en superposición de ∣0⟩ y ∣1⟩, la CNOT fuerza a q1 a un estado **no relacionado** con ∣ψ⟩, corrompiendo el objetivo. |
| Prueba final           | Comparar la posición final de q0 (post-CNOT) con la de q1 en la esfera de Bloch.             | Ninguno de los qubits está en el estado original ∣ψ⟩. El estado se transforma en un sistema entrelazado único. |

### 4. Plan de lección ejemplar: Fallo del clon cuántico universal

**Módulo:** Laboratorio aplicado 2  
Esta lección utiliza el laboratorio descrito en el prompt inicial, colocándolo aquí para finalizar la prueba empírica del teorema.

#### Laboratorio Composer: Fallo del clon cuántico universal

| Objetivo               | Probar empíricamente el Teorema de No-Clonación demostrando visualmente el fallo de la compuerta CNOT para copiar un estado arbitrario. |
| Recursos requeridos    | IBM Quantum Composer, Herramienta de visualización de la esfera de Bloch, Hoja de trabajo Nivel 4 (para registrar configuraciones RY/P) |

**Instrucciones paso a paso**

**Parte 1: Preparación del estado (∣ψ⟩)**
1. **Inicializa el lienzo:** Comienza con dos qubits, q0 (entrada) y q1 (objetivo), ambos en ∣0⟩.
2. **Define ∣ψ⟩:** Prepara q0 en un estado no base arbitrario (ejemplo: RY(π/3) seguido de P(π/2)).
3. **Verificación visual:** Usa la esfera de Bloch para confirmar la ubicación de q0. q1 debe permanecer en el Polo Norte (∣0⟩).

**Parte 2: Intento de clonación (fallo visual)**
1. **La copia ingenua:** Aplica una compuerta **CNOT** (q0 → q1).
2. **Observa el fallo:** Crucialmente, no agregues compuertas de medición al inicio. El visor de la esfera de Bloch mostrará la representación del estado entrelazado resultante.
   - **Prueba visual:** Observa q0 y q1. Si la CNOT hubiera funcionado, ambas esferas mostrarían la misma posición que el estado inicial ∣ψ⟩.
   - **Resultado esperado:** La esfera de Bloch para q1 mostrará un estado corrupto o indefinible (a menudo colapsado al origen/centro si el estado de entrada está en superposición), probando que no es una copia.
3. **Confirmación (medición):** Agrega compuertas de medición. Las probabilidades 50/50 confirman la naturaleza entrelazada del estado (como se derivó en la Semana 2), pero la esfera de Bloch proporciona la prueba visual inmediata de la corrupción.

### 5. Conclusión y siguientes pasos

Este módulo de **Nivel 4, Semana 3** concluye exitosamente la base teórica y empírica para la seguridad de la información cuántica. La confirmación visual de la corrupción del estado (fallo de clonación) sirve como principio central de seguridad.

La siguiente fase se enfocará en **Nivel 4, Semana 4** (Protocolos), donde los estudiantes usarán el estado de Bell (∣Φ+⟩) que aprendieron a crear en la Semana 2 como el **recurso necesario** para superar la limitación de No-Clonación mediante **Teleportación Cuántica** y **Codificación Superdensa**.
