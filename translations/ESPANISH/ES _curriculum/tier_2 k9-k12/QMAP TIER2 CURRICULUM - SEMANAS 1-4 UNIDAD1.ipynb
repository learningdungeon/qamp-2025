
# **Guía del Maestro: Alfabetización Cuántica y Modelado Aplicado (Tier 2)**

**Duración:** 4 semanas (aprox. 4 sesiones de 45-60 minutos cada una)  
**Audiencia:** Estudiantes de secundaria (Grados 9–12)  
**Herramienta:** Simuladores IBM Qiskit AER

---

##  **Resumen del Currículo**

- **Principio de diseño:** Alineación interdisciplinaria (Matemáticas, Ciencias y Computación).
- **Progresión de aprendizaje:**  
  1. Pre-carga conceptual (fenómenos reales)  
  2. Modelado aplicado (ecuaciones matemáticas)  
  3. Lógica computacional (Qiskit-Python)

---

##  **Secuencia Semanal**

| Semana | Tema Principal | Concepto Cuántico Clave |
|--------|---------------|-------------------------|
| 1 | Vectores de Estado y tu primer circuito Qiskit | Modelado de brújula, rotación de estado, ruido |
| 2 | De ángulos a probabilidades | Esfera de Bloch, transformaciones, rotaciones y medición |
| 3 | Entrelazamiento básico | Puertas X, CNOT, Hadamard, dos flechas magnéticas |
| 4 | Teleportación cuántica | Protocolo de teleportación, estados de Bell, correlación |

---

##  **Objetivos por Área**

- **Ciencias/Literacidad:** Definir un vector de estado cuántico, comprender superposición y analizar entrelazamiento.
- **Matemáticas:** Aplicar la regla de probabilidad total (α% + β% = 100%) para resolver ecuaciones básicas.
- **Computación:** Secuenciar comandos cuánticos (Hadamard, medición) y comprender entrelazamiento/correlación.

---

##  **Metáforas y Conceptos Fundamentales**

- **Brújula en campo magnético 2D:** Qubit, medición, superposición (el qubit puede estar en 0 y 1 a la vez).
- **Brújula en campo magnético 3D:** Esfera de Bloch, rotaciones, visualización de estados puros y aleatorios.
- **Dos brújulas 3D entrelazadas:** Entrelazamiento, compartir fase entre qubits.
- **Flecha de la brújula teleportada:** El electrón como onda de energía cuántica, comparación partícula vs. onda.
- **Superposición y spin:** Aplicación a la velocidad de las computadoras cuánticas.

---

##  **Lógica Computacional y Geometría**

- **Semanas 2-3:** Representaciones cartesianas y polares en 2D y 3D para rotación y entrelazamiento.
- **Puertas cuánticas:**  
  - Hadamard: definición de brújula  
  - RX, RY: orientación de brújula  
  - Esfera de Bloch: coordenadas formales

---

##  **Entrelazamiento y Teleportación**

- **Semana 3-4:** Introducción al entrelazamiento sin matemáticas complejas.
- **Puerta CNOT:** Entrelazamiento de dos brújulas; si el qubit 0 es “arriba”, el qubit 1 también.
- **Meta:** Teleportar el estado de la flecha de la brújula usando el protocolo de Bell.

---

##  **Recursos para Implementación**

- **Jupyter Notebook:** Plataforma principal con ejemplos de código Python/Qiskit.
- **IBM Qiskit Classroom:** Portal educativo con material y tutoriales oficiales.
- **Documentación Qiskit:** Referencia para definición de puertas y solución de problemas.

---

##  **Conclusión y Siguientes Pasos**

Este currículo de Tier 2 conecta la ciencia cuántica con la educación secundaria usando metáforas y actividades prácticas. El siguiente paso será desarrollar el puente conceptual hacia Tier 3, introduciendo formalmente la notación de Dirac y avanzando hacia temas como QAOA, VQC y criptografía cuántica.
