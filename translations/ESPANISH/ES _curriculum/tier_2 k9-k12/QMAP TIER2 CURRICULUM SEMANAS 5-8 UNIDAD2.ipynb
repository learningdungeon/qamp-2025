
# **Guía del Maestro: TOY QAOA, VQE y BB84 (Tier 2, Unidad 2)**

**Duración:** 4 semanas (aprox. 4 sesiones de 45-60 minutos cada una)  
**Audiencia:** Estudiantes de secundaria (Grados 9–12)  
**Herramienta:** Simuladores IBM Qiskit AER

---

##  **Resumen del Currículo**

- **Principio de diseño:** Alineación interdisciplinaria (Matemáticas, Ciencias y Computación).
- **Progresión de aprendizaje:**  
  1. Pre-carga conceptual (fenómenos reales)  
  2. Modelado aplicado (ecuaciones matemáticas)  
  3. Lógica computacional (Qiskit-Python)

---

##  **Secuencia Semanal**

| Semana | Tema Principal | Concepto Cuántico Clave |
|--------|---------------|-------------------------|
| 5 | Teleportación / Demostración de errores | Circuitos multiqubit, corrección y detección de errores |
| 6 | QAOA (intuitivo, bajo costo) | Función de costo, parametrización, camino más barato |
| 7 | VQE (intuitivo, brújula sacudida) | Hamiltoniano, minimización de energía, correlaciones |
| 8 | Encriptación cuántica (BB84) | Bases de medición rectilínea (Z) y diagonal (X), detección de espías |

---

##  **Objetivos por Área**

- **Ciencias/Literacidad:** Definir un vector de estado cuántico, comprender superposición y analizar entrelazamiento.
- **Matemáticas:** Aplicar la regla de probabilidad total (α% + β% = 100%) para resolver ecuaciones básicas.
- **Computación:** Secuenciar comandos cuánticos (Hadamard, medición), comprender entrelazamiento/correlación y función de costo.

---

##  **Metáforas y Conceptos Fundamentales**

- **Equilibrio de un qubit único:** Introducción a cadenas de compuertas multiqubit.
- **Brújula sacudida y equilibrada:** El Hamiltoniano como operador de energía, equilibrio y caminos de bajo costo.
- **Dos brújulas 3D entrelazadas:** Bases Z y X, detección de mensajes espiados.

---

##  **Lógica Computacional y Geometría**

- **Semanas 5-6:** Equilibrio de qubit único, comprensión del entrelazamiento.
  - Puerta CZ: entrelazamiento de brújulas, correlación cuántica.
  - Rz: mezcla de componentes ZZ en QAOA.
- **Semanas 7-8:** Modelos TOY para VQE y BB84.
  - Ansatz: múltiples configuraciones de spin, correlaciones entre vecinos.
  - Puertas Controladas-Z (CZ): correlaciones necesarias para VQE.

---

##  **Puente Conceptual Tier 1 a Tier 2**

- Reconocimiento de circuitos con 3+ qubits y el concepto de ruido como “estática”.
- Puertas controladas (Cn(U)), OpenQASM, uso de la matriz de densidad (ρ).
- Analogía: una bola asentándose en el fondo de un tazón (estado fundamental).
- Formalismo BB84: descripción formal usando matrices de rotación de bases.
