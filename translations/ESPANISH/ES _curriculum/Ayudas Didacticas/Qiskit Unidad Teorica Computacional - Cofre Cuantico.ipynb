
# **Unidad de Alfabetización Computacional Qiskit: La Bóveda Cuántica**

**Nivel objetivo:** 6° grado  
**Diseñadora de la unidad:** Noor  
**Proyecto Qiskit:** Currículo Cuántico para Todos

---

##  **Parte 1: Plan de Unidad y Marco Pedagógico**

### **Meta de la Unidad y Objetivos de Aprendizaje**

El objetivo es que los estudiantes desarrollen una comprensión básica del modelo cuántico, reconociendo su necesidad para resolver los **defectos profundos** de la física clásica, y que apliquen este conocimiento en programación cuántica conceptual.

| Área de enfoque | Objetivo (El estudiante será capaz de...) | Nivel de Bloom |
|-----------------|------------------------------------------|----------------|
| Ciencias/Literacidad | Definir y comparar los modelos atómicos clásico (Bohr) y cuántico (Onda). | Analizar, Comprender |
| Matemáticas | Aplicar la regla de probabilidad total (α% + β% = 100%) para resolver ecuaciones algebraicas básicas. | Aplicar |
| Computacional | Secuenciar comandos cuánticos conceptuales (Puerta Hadamard, Medición) para diseñar un circuito cuántico simple. | Crear |

---

### **Metodología: Flujo de la Lección y Cambio Conceptual**

La secuencia utiliza un **inicio kinestésico** para establecer la base de la unidad.

| Tiempo | Fase | Acción / Guía del docente | Enfoque VAK |
|--------|------|--------------------------|-------------|
| 5 min | El Gancho de la Burbuja Cuántica | Acción: Los estudiantes observan burbujas. Guía: "El electrón ES una Onda de Energía Cuántica. Esta burbuja modela su 'difusión' y 'no-localidad'." | Kinestésico, Visual |
| 10 min | Conflicto Conceptual | Acción: El docente presenta el modelo Bohr/Sistema Solar y sus defectos. Guía: Reconocer el "defecto profundo" (problema de colapso). Introducir la Onda Cuántica como solución estable. | Auditivo, Analítico |
| 5 min | Enlace de Medición | Acción: El docente revienta intencionalmente una burbuja. Guía: "La única vez que la onda estable colapsa es cuando usamos intencionalmente el comando de Medición para forzar una respuesta definida." | Kinestésico, Auditivo |

##  **Parte 2: Hojas de Trabajo por Materia (Asignaciones para el Estudiante)**

Estas tareas deben ser entregadas y calificadas por el docente correspondiente.

| Materia | Título de la asignación | Habilidad principal evaluada |
|---------|------------------------|-----------------------------|
| Ciencias | Hoja de trabajo: Comparando los modelos atómicos | Categorizar conceptos (Partícula vs. Onda) usando un diagrama de Venn (análisis conceptual). |
| Inglés/Literacidad | Hoja de trabajo: Narrativa de la Bóveda Cuántica | Comprensión lectora, razonamiento inferencial y adquisición de vocabulario. |
| Matemáticas | Hoja de trabajo: Álgebra y Probabilidad | Resolver ecuaciones lineales para una variable X, usando la regla de probabilidad α% + β% = 100%. |
| Computación | Hoja de trabajo: Lógica Computacional Qiskit | Comprender estados de qubit, secuenciar puertas Hadamard y Medición, y predecir resultados binarios (pensamiento computacional). |
