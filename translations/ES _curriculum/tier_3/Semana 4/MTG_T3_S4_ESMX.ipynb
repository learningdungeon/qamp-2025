## 3. Refinamientos de Lógica Computacional (Semana 4)

### A. La Preparación: Sistema de 3 Qubits

| Concepto           | Explicación                                                                 | Descripción Matemática   |
|--------------------|-----------------------------------------------------------------------------|--------------------------|
| El Gato ($q_C$)    | El estado desconocido a transmitir                                          | $[\psi\rangle = \alpha[0\rangle + \beta[1\rangle$ |
| El Canal ($q_A$, $q_B$) | El par de Bell entrelazado precompartido que conecta a Alice y Bob.         | $[\Phi^+\rangle = \frac{1}{\sqrt{2}}([00\rangle + [11\rangle)$ |

### B. El Proceso: Medición de Bell

| Concepto           | Explicación                                                                 | Descripción Matemática   |
|--------------------|-----------------------------------------------------------------------------|--------------------------|
| Cambio de base     | Alice transforma sus dos qubits ($q_A$, $q_C$) de la base computacional a la base de Bell. | $H(q_C) \cdot CNOT(q_C, q_A)$ |
| Medición           | Alice mide $q_C$ y $q_A$, colapsando el sistema en uno de 4 resultados clásicos (00, 01, 10, 11). | Medir $[q_C q_A\rangle$ |

### C. La Corrección: Feed-forward

| Concepto           | Explicación                                                                 | Descripción Matemática   |
|--------------------|-----------------------------------------------------------------------------|--------------------------|
| Bits clásicos      | Los resultados de medición deben enviarse a Bob por un canal clásico (límite de velocidad de la luz). | Enviar bits $[q_C q_A\rangle$ |
| Compuertas de corrección | Bob aplica compuertas según los bits para recuperar $[\psi\rangle$                | —                        |

## Laboratorio de Programación: Teleportación

| Objetivo           | Los estudiantes construirán el circuito de teleportación cuántica en Qiskit, verificando que el estado $|\psi\rangle$ desaparece del qubit 0 y reaparece en el qubit 2. |
|--------------------|-------------------------------------------------------------------------------------------------------------------------------------------------------------|
| Recursos requeridos| Entorno Python (Jupyter), Tier3W4_coding.ipynb, T4W4_draft.ipynb (Notas de clase)                                                                           |

### Instrucciones Paso a Paso

**Parte 1: Matemáticas (Papel y lápiz - Notas de clase)**
1. **Expansión del estado:** Seguir la derivación en T4W4_draft.ipynb. Escribir el estado completo $|\psi\rangle \otimes |\Phi^+\rangle$.
2. **Rastreo de compuertas:** Aplicar CNOT y luego H. Reagrupar términos para ver la dependencia del estado de Bob.
3. **Predicción:** Probar que si Alice mide “11”, Bob tiene el estado $ZX|\psi\rangle$.

**Parte 2: Código (Implementación en Qiskit)**
1. **Tarea 1 (Preparación):** Inicializar un circuito de 3 qubits. Preparar un estado aleatorio en $q_0$ (o uno conocido como $|1\rangle$ o $|+\rangle$). Crear un par de Bell en $q_1$, $q_2$.
2. **Tarea 2 (Caja de Alice):** Implementar la medición de Bell: `qc.cx(0, 1)` luego `qc.h(0)`. Medir $q_0$, $q_1$ en bits clásicos.
3. **Tarea 3 (Corrección de Bob):** Usar comandos `c_if`.
   - `qc.x(2).c_if(cr, 1)` (aplica X si el bit es 1)
   - `qc.z(2).c_if(cr, 2)` (aplica Z si el bit es 2; revisar el indexado de bits en Qiskit)
4. **Verificación:** Medir el qubit de Bob. Si teleportamos $|1\rangle$, Bob debe medir 1 el 100% de las veces. Si teleportamos $|+\rangle$, Bob debe medir 0 o 1 al 50%.

**Parte 3: Evaluación**
- Pregunta 3: ¿Qué información envía Alice a Bob? (Respuesta: 2 bits clásicos)
- Pregunta 5: ¿Por qué esto no es comunicación más rápida que la luz? (Respuesta: Bob necesita los bits clásicos para corregir su estado)
- Pregunta 7: ¿Qué corrección se aplica para la medición “10”? (Respuesta: compuerta Z)
