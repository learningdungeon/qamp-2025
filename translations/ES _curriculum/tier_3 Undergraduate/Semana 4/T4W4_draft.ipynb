
# Nivel 3, Semana 4: Derivación Formal de la Teleportación Cuántica

**Audiencia:** Licenciatura / Avanzado

**Objetivo:** Proporcionar una derivación matemática formal y paso a paso del protocolo de teleportación cuántica. Este mini-proyecto sintetiza todos los conceptos de Nivel 3 vistos hasta ahora (vectores de estado, compuertas unitarias, productos tensoriales y entrelazamiento).

---

## 1. Puente con Semanas Previas

La teleportación cuántica es el resumen perfecto de nuestro estudio de sistemas de uno y dos qubits. Es un protocolo que *usa* todos los conceptos que hemos dominado:

1. **Semana 1 (Vectores de Estado):** El objetivo es transmitir un estado cuántico arbitrario y desconocido (un vector de estado) $|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$ de un lugar a otro.
2. **Semana 3 (Entrelazamiento y Productos Tensoriales):** El protocolo es *imposible* sin un estado de Bell entrelazado precompartido, $|\Phi^+\rangle$. Todo el cálculo se realiza siguiendo el producto tensorial del sistema de 3 qubits.
3. **Semana 2 (Compuertas Unitarias):** El paso final del protocolo depende de que Bob aplique compuertas condicionales $X$ y $Z$ (que son rotaciones $R_x(\pi)$ y $R_z(\pi)$) para recuperar el estado original.

**Lo que la teleportación ES y NO ES:**
* **ES:** Una forma de transmitir un *estado* cuántico de un lugar a otro, destruyendo el original en el proceso (por el Teorema de No-Clonación; para más información, ver Nivel 4 Semana 1).
* **NO ES:** Comunicación instantánea. El protocolo requiere enviar dos bits clásicos de información. Esta comunicación clásica está limitada por la velocidad de la luz, así que la transferencia completa del estado no es más rápida que la luz.

---

## 2. Preparación y Estado Inicial

Tenemos un sistema de 3 qubits. Etiquetemos los qubits:
* **$q_C$ (Gato):** El estado desconocido $|\psi\rangle$ que Alice quiere enviar.  
  (*RECUERDA* que el estado a teleportar es información, no un gato físico para transportar)  
  $|\psi\rangle_C = \alpha|0\rangle + \beta|1\rangle$
* **$q_A$ (Alice):** La mitad de Alice de un par de Bell precompartido.
* **$q_B$ (Bob):** La mitad de Bob del par de Bell precompartido.

Alice tiene $q_C$ y $q_A$. Bob tiene $q_B$ y está lejos.

El estado de Bell precompartido es $|q_A q_B\rangle = |\Phi^+\rangle_{AB} = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$.

El estado inicial total del sistema de 3 qubits, $|\Psi_{total}\rangle$, es el producto tensorial del qubit de carga de Alice y el par de Bell:

$$ 
|\Psi_{total}\rangle = |\psi\rangle_C \otimes |\Phi^+\rangle_{AB} 
$$ 

Expandamos esto (el orden de los qubits será $C, A, B$):

$$ 
|\Psi_{total}\rangle = (\alpha|0\rangle_C + \beta|1\rangle_C) \otimes \frac{1}{\sqrt{2}}(|00\rangle_{AB} + |11\rangle_{AB}) 
$$ 

$$ 
|\Psi_{total}\rangle = \frac{1}{\sqrt{2}} \left[ \alpha|0\rangle_C \otimes (|00\rangle_{AB} + |11\rangle_{AB}) + \beta|1\rangle_C \otimes (|00\rangle_{AB} + |11\rangle_{AB}) \right] 
$$ 

$$ 
|\Psi_{total}\rangle = \frac{1}{\sqrt{2}} \left[ \alpha(|000\rangle + |011\rangle) + \beta(|100\rangle + |111\rangle) \right] 
$$ 

Este es nuestro punto de partida. Alice tiene los dos primeros qubits; Bob tiene el tercero.

---

## 3. Operaciones de Alice: Medición en la Base de Bell

Alice no puede simplemente medir $|\psi\rangle$—eso lo destruiría. En cambio, realiza una "Medición en la Base de Bell" en sus dos qubits ($q_C$ y $q_A$). Esto implica dos compuertas:

1. **Compuerta $CNOT$:** Alice aplica un $CNOT$ con $q_C$ como control y $q_A$ como objetivo.
2. **Compuerta Hadamard:** Alice aplica una compuerta $H$ a $q_C$.

Sigamos el vector de estado a través de estas dos operaciones.

### Paso A: Después del $CNOT$ de Alice (control: $q_C$, objetivo: $q_A$)

El $CNOT$ invierte $q_A$ si $q_C$ es 1. Aplicamos esto a nuestros cuatro términos:
* $\alpha|000\rangle \to \alpha|000\rangle$ (Sin cambio)
* $\alpha|011\rangle \to \alpha|011\rangle$ (Sin cambio)
* $\beta|100\rangle \to \beta|110\rangle$ ($q_A$ cambia de 0 a 1)
* $\beta|111\rangle \to \beta|101\rangle$ ($q_A$ cambia de 1 a 0)

El estado ahora es $|\Psi_1\rangle$:
$$ 
|\Psi_1\rangle = \frac{1}{\sqrt{2}} \left[ \alpha(|000\rangle + |011\rangle) + \beta(|110\rangle + |101\rangle) \right] 
$$ 

### Paso B: Después del Hadamard de Alice (en $q_C$)

Ahora Alice aplica $H$ a $q_C$ (el primer qubit).
* $|0\rangle \to H|0\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$
* $|1\rangle \to H|1\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)$

Aplicamos esto a los cuatro términos en $|\Psi_1\rangle$:
$$ 
|\Psi_2\rangle = \frac{1}{\sqrt{2}} \left[ \alpha\left(\frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)|00\rangle + \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)|11\rangle \right) + \beta\left(\frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)|10\rangle + \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)|01\rangle \right) \right] 
$$ 

Expandimos y sacamos el $\frac{1}{\sqrt{2}}$ al frente (el prefactor total es $\frac{1}{2}$):
$$ 
|\Psi_2\rangle = \frac{1}{2} \left[ \alpha(|000\rangle + |100\rangle + |011\rangle + |111\rangle) + \beta(|010\rangle - |110\rangle + |001\rangle - |101\rangle) \right] 
$$

---

## 4. El "Truco": Reagrupando por los Estados Base de Alice

Este es el paso más crítico de la derivación. El estado $|\Psi_2\rangle$ parece un desastre. Pero vamos a *reagrupar* los términos según lo que controla Alice ($q_C, q_A$) para ver qué ha pasado con el qubit de Bob ($q_B$).

Vamos a factorizar los cuatro posibles estados base de Alice: $|00\rangle_{CA}$, $|01\rangle_{CA}$, $|10\rangle_{CA}$, $|11\rangle_{CA}$.

$$ 
|\Psi_2\rangle = \frac{1}{2} [ \alpha|000\rangle+\beta|001\rangle+\alpha|011\rangle+\beta|010\rangle+\alpha|100\rangle-\beta|101\rangle+\alpha|111\rangle-\beta|110\rangle ]
$$ 

* **$|00\rangle_{CA}$:** (Busca todos los términos que empiezan con $|00\rangle$)
    * $\alpha|000\rangle \to \alpha|0\rangle_B$
    * $\beta|001\rangle \to \beta|1\rangle_B$
    * $\implies |00\rangle_{CA} \otimes (\alpha|0\rangle + \beta|1\rangle)_B$

* **$|01\rangle_{CA}$:** (Busca todos los términos que empiezan con $|01\rangle$)
    * $\alpha|011\rangle \to \alpha|1\rangle_B$
    * $\beta|010\rangle \to \beta|0\rangle_B$
    * $\implies |01\rangle_{CA} \otimes (\beta|0\rangle + \alpha|1\rangle)_B$

* **$|10\rangle_{CA}$:** (Busca todos los términos que empiezan con $|10\rangle$)
    * $\alpha|100\rangle \to \alpha|0\rangle_B$
    * $-\beta|101\rangle \to -\beta|1\rangle_B$
    * $\implies |10\rangle_{CA} \otimes (\alpha|0\rangle - \beta|1\rangle)_B$

* **$|11\rangle_{CA}$:** (Busca todos los términos que empiezan con $|11\rangle$)
    * $\alpha|111\rangle \to \alpha|1\rangle_B$
    * $-\beta|110\rangle \to -\beta|0\rangle_B$
    * $\implies |11\rangle_{CA} \otimes (-\beta|0\rangle + \alpha|1\rangle)_B$

¡Este estado reagrupado es la clave de todo el protocolo!

---

## 5. Medición de Alice y Corrección de Bob

Ahora Alice mide sus dos qubits, $q_C$ y $q_A$. Su medición (en la base Z) será $00$, $01$, $10$ o $11$, cada una con 25% de probabilidad. Esta medición *colapsa* todo el estado de 3 qubits en uno de los cuatro términos que acabamos de reagrupar.

El qubit de Bob, $q_B$, ahora está en un nuevo estado, *condicional* a la medición de Alice.

**Caso 1: Alice mide $00$**
* El sistema colapsa a: $\frac{1}{2} |00\rangle_{CA} \otimes (\alpha|0\rangle + \beta|1\rangle)_B$
* El qubit de Bob es: $(\alpha|0\rangle + \beta|1\rangle)$. **¡Este es exactamente $|\psi\rangle$!**
* Alice envía "00" a Bob. Bob no hace nada (aplica una compuerta $I$).

**Caso 2: Alice mide $01$**
* El sistema colapsa a: $\frac{1}{2} |01\rangle_{CA} \otimes (\beta|0\rangle + \alpha|1\rangle)_B$
* El qubit de Bob es: $(\beta|0\rangle + \alpha|1\rangle)$. Es $|\psi\rangle$ pero con los coeficientes intercambiados.
* Alice envía "01" a Bob. Bob aplica una **compuerta Pauli-$X$**:
    * $X (\beta|0\rangle + \alpha|1\rangle) = \beta|1\rangle + \alpha|0\rangle = \alpha|0\rangle + \beta|1\rangle = |\psi\rangle$.
    * Esta compuerta Pauli-$X$ asegura que la información de $|\psi\rangle$ se transfiere exactamente a Bob.

**Caso 3: Alice mide $10$**
* El sistema colapsa a: $\frac{1}{2} |10\rangle_{CA} \otimes (\alpha|0\rangle - \beta|1\rangle)_B$
* El qubit de Bob es: $(\alpha|0\rangle - \beta|1\rangle)$. Es $|\psi\rangle$ pero con un cambio de fase en $|1\rangle$.
* Alice envía "10" a Bob. Bob aplica una **compuerta Pauli-$Z$**:
    * $Z (\alpha|0\rangle - \beta|1\rangle) = \alpha|0\rangle - (-\beta|1\rangle) = \alpha|0\rangle + \beta|1\rangle = |\psi\rangle$.

**Caso 4: Alice mide $11$**
* El sistema colapsa a: $\frac{1}{2} |11\rangle_{CA} \otimes (-\beta|0\rangle + \alpha|1\rangle)_B$
* El qubit de Bob es: $(-\beta|0\rangle + \alpha|1\rangle)$. Este estado está intercambiado *y* con cambio de fase.
* Alice envía "11" a Bob. Bob aplica primero una **compuerta $X$**, luego una **compuerta $Z$**:
    * **Primero $X$:** $X(-\beta|0\rangle + \alpha|1\rangle) = -\beta|1\rangle + \alpha|0\rangle$
    * **Luego $Z$:** $Z(\alpha|0\rangle - \beta|1\rangle) = \alpha|0\rangle - (-\beta|1\rangle) = \alpha|0\rangle + \beta|1\rangle = |\psi\rangle$.

---

## 6. Conclusión

En los cuatro posibles resultados, Bob puede reconstruir perfectamente el estado original $|\psi\rangle$ aplicando una corrección específica (I, X, Z o ZX) según los dos bits clásicos que recibe de Alice.

El protocolo logra "teleportar" el estado desconocido $|\psi\rangle$ de Alice a Bob, transfiriendo la información (estado del qubit) a Bob. Este protocolo no está limitado por la distancia entre Alice y Bob; el estado del qubit colapsa inmediatamente en el lado de Bob cuando Alice mide y colapsa su propio qubit ($q_A$). Sin embargo, la información no viaja instantáneamente, ya que Alice debe decirle a Bob qué compuerta aplicar, por lo que la información no viaja a la velocidad de la luz ni más rápido.

En el proceso, el estado original de Alice se destruye por la medición, cumpliendo el Teorema de No-Clonación. Esta derivación muestra cómo el entrelazamiento sirve como recurso cuántico, activado por comunicación clásica, para lograr una tarea imposible en la física clásica.
